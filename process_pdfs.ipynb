## Travis Co. 2020

```python
pcts = []
trumptotals = []
bidentotals = []
cornyntotals=[]
hegartotals = []
davistotals = []
roytotals = []
travislines = open('partial-counties/travis/travis-2020final.txt').readlines()
dropindices = []
tx10pcts =[]
tx21pcts =[]
tx25pcts =[]
tx17pcts= []
tx35pcts = []
for i in range(len(travislines)):
    if str(travislines[i])=='\n':
        dropindices.append(i)
for index in sorted(dropindices, reverse=True):
    del travislines[index]
for i in range(len(travislines)):
    #print(travislines[i])
    try:
        if 'Travis County, Texas' in str(travislines[i]):
            #print(travislines[i-5:i+5])
            if str(travislines[i+1]).strip() not in pcts:
                pcts.append(str(travislines[i+1]).strip())
        elif 'Donald J. Trump' in str(travislines[i]):
            trumptotals.append(int(travislines[i+1].strip().replace(',','')))
        elif 'Joseph R. Biden' in str(travislines[i]):
            bidentotals.append(int(travislines[i+1].strip().replace(',','')))
        elif 'United States Representative District 10' in str(travislines[i]):
            tx10pcts.append(str(travislines[i:][travislines[i:].index('Travis County, Texas\n')+1]).strip())
        elif 'United States Representative District 21' in str(travislines[i]):
            tx21pcts.append(str(travislines[i:][travislines[i:].index('Travis County, Texas\n')+1]).strip())
        elif 'United States Representative District 25' in str(travislines[i]):
            tx25pcts.append(str(travislines[i:][travislines[i:].index('Travis County, Texas\n')+1]).strip())
        elif 'United States Representative District 17' in str(travislines[i]):
            tx17pcts.append(str(travislines[i:][travislines[i:].index('Travis County, Texas\n')+1]).strip())
        elif 'United States Representative District 35' in str(travislines[i]):
            tx35pcts.append(str(travislines[i:][travislines[i:].index('Travis County, Texas\n')+1]).strip())
        elif "John Cornyn" in str(travislines[i]):
            try:
                cornyntotals.append(int(travislines[i+1].strip().replace(',','')))
            except:
                try:
                    cornyntotals.append(int(travislines[i+8].strip().replace(',','')))
                except:
                    cornyntotals.append(int(travislines[i+7].strip().replace(',','')))
        elif """Mary "MJ" Hegar""" in str(travislines[i]):
            try:
                if '%' in str(travislines[i+2]) or str(travislines[i+2]) == '0\n':
                    hegartotals.append(int(travislines[i+1].strip().replace(',','')))
                else:
                    raise TypeError('bad')
            except:
                try:
                    if '%' in str(travislines[i+14]):
                        hegartotals.append(int(travislines[i+13].strip().replace(',','')))
                    else:
                        raise TypeError('bad')
                except:
                    try:
                        if '%' in str(travislines[i+13]):
                            hegartotals.append(int(travislines[i+12].strip().replace(',','')))
                        else:
                            raise TypeError('bad')
                    except:
                        try:
                            if '%' in str(travislines[i-26]):
                                hegartotals.append(int(travislines[i-27].strip().replace(',','')))
                            else:
                                raise TypeError('bad')
                        except:
                            if '%' in str(travislines[i-27]):
                                hegartotals.append(int(travislines[i-28].strip().replace(',','')))
                            else:
                                raise TypeError('bad')
        elif "Wendy R. Davis" in str(travislines[i]):
            try:    
                if '%' in str(travislines[i+2]) or str(travislines[i+2]) == '0\n':
                    davistotals.append(int(travislines[i+1].strip().replace(',','')))
                else:
                    raise TypeError('bad')
            except:
                try:
                    if '%' in str(travislines[i-21]) or str(travislines[i-21]) == '0\n':
                        davistotals.append(int(travislines[i-22].strip().replace(',','')))
                    else:
                        raise TypeError('bad')
                except:
                    try:
                        if '%' in str(travislines[i+11]) or str(travislines[i+11]) == '0\n':
                            davistotals.append(int(travislines[i+10].strip().replace(',','')))
                        else:
                            raise TypeError('bad')
                    except:
                        if '%' in str(travislines[i+13]) or str(travislines[i+13]) == '0\n':
                            davistotals.append(int(travislines[i+12].strip().replace(',','')))
                        else:
                            raise TypeError('bad')
                        
        elif "Chip Roy" in str(travislines[i]):
            try:    
                if '%' in str(travislines[i+2]) or str(travislines[i+2]) == '0\n':
                    roytotals.append(int(travislines[i+1].strip().replace(',','')))
                else:
                    raise TypeError('bad')
            except:
                try:
                    if '%' in str(travislines[i+8]) or str(travislines[i+8]) == '0\n':
                        roytotals.append(int(travislines[i+7].strip().replace(',','')))
                    else:
                        raise TypeError('bad')
                except:
                    try:
                        if '%' in str(travislines[i+7]) or str(travislines[i+7]) == '0\n':
                            roytotals.append(int(travislines[i+6].strip().replace(',','')))
                        else:
                            raise TypeError('bad')
                    except:
                        if '%' in str(travislines[i+9]) or str(travislines[i+9]) == '0\n':
                            roytotals.append(int(travislines[i+8].strip().replace(',','')))
                        else:
                            raise TypeError('bad')
    except:
        print(pcts[-1],travislines[i])
        display(travislines[i-30:i+30])
        break
print(len(hegartotals),len(cornyntotals))
print(len(trumptotals),len(bidentotals))
print(len(davistotals),len(roytotals),len(tx21pcts))
```

# Bexar Co. 2020

```python
text = open('partial-counties/bexar/bexar-2020all.txt')
biden = {}
trump={}
pcts = {}
statepcts = {}
cornyn={}
hegar={}
for line in text:
    if 'Biden' in line:
        biden[line[7:11]] =int(line[11:17])
    elif 'Trump' in line:
        trump[line[7:11]]=int(line[11:17])
    elif 'Cornyn' in line:
        cornyn[line[7:11]]=int(line[11:17])
    elif 'Hegar' in line:
        hegar[line[7:11]]=int(line[11:17])
    elif 'United States Representative' in line:
        pcts[line[7:11]]=line[104:111].strip()[-2:]
    elif 'STR121' in line:
        print(line)
        statepcts[line[7:11]]=121
        
trump = pd.DataFrame.from_dict(trump,orient='index',columns=['Trump_2020'])
biden = pd.DataFrame.from_dict(biden,orient='index',columns=['Biden_2020'])
cornyn = pd.DataFrame.from_dict(cornyn,orient='index',columns=['Cornyn_2020'])
hegar = pd.DataFrame.from_dict(hegar,orient='index',columns=['Hegar_2020'])
pcts = pd.DataFrame.from_dict(pcts,orient='index',columns=['CD'])
statepcts = pd.DataFrame.from_dict(statepcts,orient='index',columns=['HD'])

pcts['Precinct'] = pcts.index
pcts.to_csv('partial-counties/bexar/bexar-precincts2020.csv',index=False)

statepcts['Precinct'] = statepcts.index
statepcts.to_csv('partial-counties/bexar/bexar-stateprecincts2020.csv',index=False)

bexar2020p = trump.merge(biden,on=biden.index)
bexar2020p['Precinct']= bexar2020p['key_0']
bexar2020p = bexar2020p.drop(columns='key_0')
bexar2020p.to_csv('partial-counties/bexar/bexar-2020president.csv',index=False)

bexar2020s = hegar.merge(cornyn,on=cornyn.index)
bexar2020s['Precinct']= bexar2020s['key_0']
bexar2020s = bexar2020s.drop(columns='key_0')
bexar2020s.to_csv('partial-counties/bexar/bexar-2020senate.csv',index=False)
```
